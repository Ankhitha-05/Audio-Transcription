```python
import gradio as gr
from whisper import load_model

# Load Whisper model
model = load_model("base")

def transcribe_audio(audio, language):
    if audio is None:
        return "Please upload an audio file."

    # Load audio and transcribe
    result = model.transcribe(audio, language=language)
    return result["text"]

with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("<h1>🎙 Multi-Language Audio Transcription</h1>")

    with gr.Row():
        audio_input = gr.Audio(label="Upload Audio", type="filepath")
        lang_input = gr.Dropdown(
            ["English", "Spanish", "French", "Hindi", "Arabic"],
            label="Select Language",
            value="English"
        )

    output_text = gr.Textbox(label="Transcription Output")

    btn = gr.Button("Transcribe", variant="primary")
    btn.click(transcribe_audio, inputs=[audio_input, lang_input], outputs=output_text)

demo.launch(share=True)
```
